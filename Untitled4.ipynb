```python
#import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pip install openpyxl
```

```python
# exploring content table
content= pd.read_csv('Social Buzz/Content.csv')
content=content.dropna()
content = content.rename(columns={'Type':'Content Type'})
clean_content= content[['Content ID','Content Type','Category']]
clean_content.info()
print(clean_content['Category'].unique())
print(clean_content.head())
```

```python
# exploring reactions table
reactions=pd.read_csv('Social Buzz/Reactions.csv')
reactions=reactions.dropna()
reactions.info()
reactions=reactions.rename(columns={'Type': 'Reactions Type'})
clean_reactions=reactions[['Content ID', 'Reactions Type']]
print(clean_reactions.head())
```

```python
# exploring reactions type table
reac_type= pd.read_csv('Social Buzz/ReactionTypes.csv')
reac_type.info()
reac_type=reac_type.rename(columns={'Type': 'Reactions Type'})
clean_reac_type=reac_type[['Reactions Type', 'Sentiment', 'Score']]
print(clean_reac_type.head())
```

```python
#merging 3 tables to one
merge_df=clean_content.merge(clean_reactions, on='Content ID').merge(clean_reac_type, on='Reactions Type')

print(merge_df.head())
```

```python
#saving excel file
merge_df.to_excel('clean_content.xlsx', index=False, engine='openpyxl')
```

```python
#exploring top 5 categoris
top_5_categories= merge_df.groupby('Category')['Score'].sum()
top_5_categories=top_5_categories.nlargest(5).reset_index()
print(top_5_categories)
# filtering data with top 5 categories
filtered_df = merge_df[merge_df['Category'].isin(top_5_categories)]

# grouping by category and counting score for sorting
category_scores_for_sorting = filtered_df.groupby('Category')['Score'].sum()

# saving new column as total score to  filtered_df
filtered_df['Total Score'] = filtered_df['Category'].map(category_scores_for_sorting)

# sorting
sorted_filtered_df = filtered_df.sort_values(by=['Total Score'], ascending=False)
print(sorted_filtered_df.head())
```

```python
#saving new excel with data of top 5 categories
sorted_filtered_df.to_excel('top_5_categories.xlsx', index=False, engine='openpyxl')
```

```python
# counting % of top 5 categories
total_score = top_5_categories['Score'].sum()
top_5_categories['Percentage'] = (top_5_categories['Score'] / total_score) * 100

# let's make beautiful colour palette
palette_color = sns.color_palette('viridis', n_colors=len(top_5_categories))

# presenting top categories in a pie chart
fig, ax = plt.subplots(figsize=(8, 8))
wedges, texts, autotexts = ax.pie(
    top_5_categories['Percentage'],
    labels=top_5_categories['Category'],
    autopct='%1.1f%%',
    colors=palette_color,
    startangle=140,
    wedgeprops={'edgecolor': 'grey'}
)

# labels
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(12)

# title
plt.title('Top 5 Content Categories by Score (Percentage)', fontsize=16)
# Show the plot
plt.show()
# Save the plot as a file (e.g., PNG)
plt.savefig('top_5_content_categories.jpg', format='jpg')
```

```python
#How many unique categories are there?
# Находим количество уникальных категорий
merge_df.drop_duplicates()
print(merge_df['Category'].value_counts())
# Объедините два столбца в один, приводя к нижнему регистру для унификации
# Convert all values in the 'Category' column to lowercase
merge_df['Category'] = merge_df['Category'].str.lower()


unique_categories = merge_df['Category'].nunique()

# Создаем график для визуализации
plt.figure(figsize=(8, 5))
sns.countplot(x='Category', data=merge_df, palette='viridis', order=merge_df['Category'].value_counts().index)
plt.title(f'Number of Unique Categories: {unique_categories}')
plt.xlabel('Category')
plt.ylabel('Quantity')
plt.xticks(rotation=90)
plt.show()
# Save the plot as a file (e.g., PNG)
plt.savefig('Number of Unique Categories.png', format='png', dpi=300, bbox_inches='tight')
```

```python
#What was the month with the most posts?
print(reactions.head())
# Convert the 'Datetime' column in Reaction to datetime format
reactions['Datetime'] = pd.to_datetime(reactions['Datetime'])

# Merge the Reaction table with Content table on Content ID
df_merged = pd.merge(reactions, clean_content, on='Content ID')

# Extract month and year from the 'Datetime' column
df_merged['Month'] = df_merged['Datetime'].dt.to_period('M')

# Count posts by month
monthly_post_counts = df_merged['Month'].value_counts()

# Find the month with the most posts
most_posts_month = monthly_post_counts.idxmax()
most_posts_count = monthly_post_counts.max()

# Convert the Series to a DataFrame for Seaborn
monthly_post_counts_df = monthly_post_counts.reset_index()
monthly_post_counts_df.columns = ['Month', 'Number of Posts']

# Plot the number of posts by month using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Month', y='Number of Posts',palette='viridis', data=monthly_post_counts_df)

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Number of Posts')
plt.title('Number of Posts by Month')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Show the plot
plt.tight_layout()
plt.show()

plt.savefig('Number of Posts by month.png', format='png', dpi=300, bbox_inches='tight')
```
